```python
# Игра 'Крестики-нолики'

import random
from random import randint

a = 10           
PLAY_BOARD = [0] * a
for i in range(10): 
  PLAY_BOARD[i] = [0] * a 

PLAYERS_MARKS = ['X', 'O']

m = 0
for i in range(10):
  for j in range(10):
    m += 1
    PLAY_BOARD[i][j] = m


# Генерация игрового поля
def display_board(board_list):
  cur_element = 0
  for i in range(10):
    for j in range(9):
      cur_element = board_list[i][j]
      if (len(str(cur_element)) > 1):
        print(cur_element, end=' | ')
      else:
        print(cur_element, end='  | ')
    print(board_list[i][j+1])


# Выбор игровой роли: крестик или нолик
def player_input():
  player_first = ''
  player_second = ''

  while player_first not in ('X', 'O'):
    player_first = input('Вы хотите играть за X или O? ').upper()

  if player_first == 'X':
    player_second = 'O'
  else:
    player_second = 'X'

  return player_first, player_second


# Установка маркера игрока в указанную позицию
def place_marker(board, marker, position):    
  if (position != 0):
    position -= 1
    i = position // 10
    j = position % 10
    board[i][j] = marker


# Проверка выиграл ли игрок с указанным маркером игру
def lose_check(board, mark):
  count = 0
  
  for i in range(10):
    count = 0
    for j in range(10):
      if (board[i][j] == mark):
        count += 1
    if (count > 4): 
      return True
  
  for i in range(10):
    count = 0
    for j in range(10):
      if (board[j][i] == mark):
        count += 1
    if (count > 4): 
      return True
  
  count = 0
  
  for i in range(10):
    if (board[i][i] == mark):
        count += 1
    if (count > 4): 
      return True

  return False


# Определение случайным образом игрока, который будет ходить первым
def choose_first():
  return PLAYERS_MARKS[random.choice((0, 1))]


# Определение пуста ли ячейка в указанной позиции
def space_check(board, position):
  i = position // 10
  j = position % 10
  if board[i][j - 1] not in PLAYERS_MARKS:
    return True


# Определяет имеется ли на игровой доске оба маркера: X и O
def full_board_check(board):
  x = len(PLAY_BOARD[0])
  y = len(PLAY_BOARD)

  lst = []
  for i in range(y):
    for j in PLAY_BOARD[i]:
      lst.append(j)
  return len(lst) == 2


# Выбор игроком следующей ячейки для хода и проверка того 
# можно ли поставить маркер в эту ячейку
def player_choice(board, player_mark):
    position = 0

    while position not in [num for num in range(1, 100)]:
      try:
        position = int(input(f'Игрок "{player_mark}", выберите ячейку с 1 по 100: '))

      except ValueError as exc:
        print(f'Неверное значение: {exc}. Пожалуйста, попробуйте снова.')

    if space_check(board, position):
      return position

    return 0


# Предложение игрокам начать игру заново
def replay():
    decision = ""
    while decision not in ('да', 'нет'):
        decision = input(
            'Вы бы хотели поиграть еще раз? Напишите "да" или "нет" '
        ).lower()

    return decision == 'да'


# Очищение игрового экрана добавлением пустых строк
def clear_screen():
    print('\n' * 100)


# Переключение роли игрока для смены очереди для хода
def switch_player(mark):
    if mark == 'X': 
      return 'O'  
    else:
      return 'X'


# Проверка того, завершена ли игра
def check_game_finish(board, mark):
    if lose_check(board, mark):
        print(f'Игрок "{mark}" выиграл!')
        return True

    if full_board_check(PLAY_BOARD):
        print('Игра завершилась вничью.')
        return True

    return False


# Поиск лучшей позиции для простановки компьютером
def find_best_position(board, CURRENT_PLAYER_MARK):
  position = 0
  best_col = 0
  best_row = 0
  count = 0
  max = 0
  
  for i in range(10):
    count = 0
    for j in range(10):
      if (board[i][j] == CURRENT_PLAYER_MARK):
        count += 1
    if (max < count):
      max = count
      best_row = i
  
  max = 0
  for i in range(10):
    count = 0
    for j in range(10):
      if (board[j][i] == CURRENT_PLAYER_MARK):
        count += 1
    if (max < count):
      max = count
      best_col = j

  if (best_row > best_col):
    for i in range(10):
      if (board[best_row][i] not in PLAYERS_MARKS):
        return (best_row * 10 + i + 1)
  else:
    for i in range(10):
      if (board[i][best_col] not in PLAYERS_MARKS):
        return (i * 10 + best_col + 1)

  for i in range(10):
      if (board[0][i] not in PLAYERS_MARKS):
        return (i + 1)

# Ход компьютера
def computer_step(board, CURRENT_PLAYER_MARK):
  position = find_best_position(board, CURRENT_PLAYER_MARK)
  place_marker(board, CURRENT_PLAYER_MARK, position)


# Приветствие пользователя
print('Добро пожаловать в игру "Крестики-нолики"!')

# Выбор пользователем игровой роли: x или 0
PLAYER_MARKS = player_input()

# Выбор случайным образом того, кто сделает первый ход(крестик или нолик)    
CURRENT_PLAYER_MARK = choose_first()

# Отображение пользователю результата выбора игрока с первым ходом
print(f'Игрок "{CURRENT_PLAYER_MARK}" ходит первым.')

# Процесс игры
while True:
    display_board(PLAY_BOARD)

    print(f'Очередь игрока "{CURRENT_PLAYER_MARK}":')

    if (CURRENT_PLAYER_MARK == PLAYER_MARKS[1]):
      computer_step(PLAY_BOARD, CURRENT_PLAYER_MARK)
    
    else:
      PLAYER_POSITION = player_choice(PLAY_BOARD, CURRENT_PLAYER_MARK)
      place_marker(PLAY_BOARD, CURRENT_PLAYER_MARK, PLAYER_POSITION)

    if check_game_finish(PLAY_BOARD, CURRENT_PLAYER_MARK):
        display_board(PLAY_BOARD)
        if not replay():
            break
        else:
            PLAY_BOARD = [0] * a
            for i in range(10): 
              PLAY_BOARD[i] = [0] * a 
            
            PLAYER_MARKS = player_input()
            CURRENT_PLAYER_MARK = choose_first()
    else:
        if (PLAYER_POSITION != 0):
          CURRENT_PLAYER_MARK = switch_player(CURRENT_PLAYER_MARK)
    clear_screen()

```
